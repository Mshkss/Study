11. У вас есть массив, представляющий собой скорости движения транспортного средства в км/ч: speeds = np.array([60, 75, 90, 100, 80, 65]).
    Создайте массив travel_times, содержащий расчетное время в пути для каждой скорости при преодолении расстояния 150 км с учетом замедления движения на 20%. Когда будете считать время, округлите  результат вниз, возвращая только целую часть.

```python
import numpy as np

speeds = np.array([60, 75, 90, 100, 80, 65])
distance = 150
slowdown_rate = 0.80

# здесь должен быть ваш код

speeds_slowed = speeds * slowdown_rate
travel_times = distance/speeds_slowed
print(speeds_slowed)

print("Время в пути для каждой скорости:", np.floor(travel_times).astype(int))

# Время в пути для каждой скорости: [3. 2. 2. 1. 2. 2.]
```

13. Используя циклы написать собственную реализацию умножения двух двумерных матриц A и B, засечь время выполнения и сравнить его с функцией умножения матриц NumPy:

```python
import time

n = 50 # значение может изменяться
A = np.random.rand(n, n)
B = np.random.rand(n, n)
C = np.zeros((n, n))
C_custom = np.zeros((n, n))
# print(A)
# print(B)
# print(C)

# Умножение матриц с использованием циклов
start_time = time.time()
for i in range(n):
    for j in range(n):
        for k in range(n):
            C_custom[i, j] += A[i, k] * B[k, j]
custom_time = time.time() - start_time

# Измерение времени выполнения NumPy
start_time = time.time()
C_numpy = np.dot(A, B)
numpy_time = time.time() - start_time

# Вывод результатов
print(f"Время выполнения собственной реализации: {custom_time:.6f} секунд")
print(f"Время выполнения NumPy: {numpy_time:.6f} секунд")

# Проверка корректности результата
print("Результаты совпадают:", np.allclose(C_custom, C_numpy))
```
